```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
pic = cv2.imread('sammy.jpg')
fix_pic = cv2.cvtColor(pic,cv2.COLOR_BGR2RGB)

print(fix_pic.shape)
width = fix_pic.shape[1]
height = fix_pic.shape[0]
print(f"width: {width}, height: {height}")

fig = plt.figure(figsize=(20,15))
rows = 3
cols = 4
ax = fig.add_subplot(rows,cols,1)
ax.imshow(fix_pic)
ax.set_title('original image')

#Show image in grayscale
gray_pic = cv2.imread('sammy.jpg', cv2.IMREAD_GRAYSCALE)
ax = fig.add_subplot(rows,cols,2)
ax.imshow(gray_pic, cmap='gray')
ax.set_title('Gray image')

#resize based on ratio
w_ratio = 0.7
h_ratio = 0.5
ratio_pic = cv2.resize(fix_pic,(0,0),fix_pic,w_ratio,h_ratio)
ax = fig.add_subplot(rows,cols,3)
ax.imshow(ratio_pic)
ax.set_title('resize based on ratio')

#resize based on selected size
resize_pic = cv2.resize(fix_pic,(500,400))
ax = fig.add_subplot(rows, cols, 4)
ax.imshow(resize_pic)
ax.set_title('resize based on selected size')

#Flip image
flip_pic = cv2.flip(fix_pic,0)
ax = fig.add_subplot(rows,cols,5)
ax.imshow(flip_pic)
ax.set_title('Flip an image')

# Draw a rectangle on image
cv2.rectangle(fix_pic, pt1=(250,350), pt2=(750,700), color=(255,0,0), thickness=5)
ax = fig.add_subplot(rows, cols, 6)
ax.imshow(fix_pic)
ax.set_title('Face recognition')

# Draw a circle on image
cv2.circle(img=fix_pic, center=(600,800), radius=50, color=(0,255,0), thickness=5)
ax = fig.add_subplot(rows,cols, 7)
ax.imshow(fix_pic)
ax.set_title('Dog color - object recognition')

# Draw a circle, fill it and draw a line
cv2.circle(img=fix_pic, center=(400,500), radius=30, color=(0,255,0), thickness=-1)
cv2.line(fix_pic, pt1=(0,0), pt2=(1025,1367), color=(255,255,255), thickness=10)
ax = fig.add_subplot(rows,cols, 8)
ax.imshow(fix_pic)
ax.set_title('Fill eyes, draw line')

# put text on image
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(gray_pic, text='Dog', org=(450,200), fontFace=font, fontScale=4, color=(255,0,0), thickness=5, lineType=cv2.LINE_AA)
ax = fig.add_subplot(rows,cols,9)
ax.imshow(gray_pic)
ax.set_title('Put Text')

# Draw custom polygon
vertices = np.array([ [500,1300], [250,500], [550,300], [900,600], [900,1100]], dtype=np.int32).reshape(-1,1,2)
cv2.polylines(gray_pic, [vertices], isClosed=True, color=(255,0,0), thickness=10)
ax = fig.add_subplot(rows,cols,10)
ax.imshow(gray_pic)
ax.set_title('Draw Polygon')


# HSV image
pic2 = cv2.cvtColor(pic, cv2.COLOR_BGR2HSV)
ax = fig.add_subplot(rows,cols,11)
ax.imshow(pic2)
ax.set_title('HSV image')
```

```python
# Two images in different size
img1 = cv2.imread('dog_backpack.jpg')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.imread('watermark_no_copy.png')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(1,4,1)
ax.imshow(img1)
ax.set_title(f"shape {img1.shape}")

ax = fig.add_subplot(1,4,2)
ax.imshow(img2)
ax.set_title(f"shape {img2.shape}")

# Blending images of the same size
img1 = cv2.resize(img1, (1200,1200))
img2 = cv2.resize(img2, (1200,1200))
blended = cv2.addWeighted(src1=img1, alpha=0.2, src2=img2,beta=0.6,gamma=0)

ax = fig.add_subplot(1,4,3)
ax.imshow(blended)
ax.set_title(f"shape {blended.shape}")

# OVERLAY SMALL IMAGE ON TOP OF A LARGER
# Numpy reassignment
img1 = cv2.imread('dog_backpack.jpg')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.imread('watermark_no_copy.png')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

img2 = cv2.resize(img2,(600,600))
large_img = img1
small_img = img2
x_offset = 0
y_offset = 0
x_end = x_offset + small_img.shape[1]
y_end = y_offset + small_img.shape[0]
large_img[y_offset:y_end, x_offset:x_end] = small_img
ax = fig.add_subplot(1,4,4)
ax.imshow(large_img)
```

```python
fig = plt.figure(figsize=(15,10))
rows = 2
cols = 4

ax = fig.add_subplot(rows, cols,1)
img1 = cv2.imread('road_image.jpg')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
ax.imshow(img1)
ax.set_title(f"Image 1:{img1.shape}")

ax = fig.add_subplot(rows, cols,2)
img2 = cv2.imread('watermark_no_copy.png')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
ax.imshow(img2)
ax.set_title(f"Image 2:{img2.shape}")

#Resize img2
img2 = cv2.resize(img2, (300,300))
ax = fig.add_subplot(rows, cols,3)
ax.imshow(img2)
ax.set_title(f"resize img 2")

gray_img = cv2.cvtColor(img2, cv2.COLOR_RGB2GRAY)
ax = fig.add_subplot(rows, cols,4)
ax.imshow(gray_img, cmap='gray')
ax.set_title(f"gray img 2")

#Inverse image 2
mask_inv = cv2.bitwise_not(gray_img)
ax = fig.add_subplot(rows, cols,5)
ax.imshow(mask_inv, cmap='gray')
ax.set_title(f"inverse img 2{mask_inv.shape}")

#Add color channel to inverse image
white_bk = np.full(img2.shape,255,dtype=np.uint8)
bk = cv2.bitwise_or(white_bk, white_bk, mask=mask_inv)
ax = fig.add_subplot(rows, cols,6)
ax.imshow(bk)
ax.set_title(f"back image")


# Replace with original image
fg = cv2.bitwise_or(img2,img2, mask=mask_inv)
ax = fig.add_subplot(rows, cols,7)
ax.imshow(fg)
ax.set_title(f"front image")

# Put mask on background image
x_offset = img1.shape[1] - 300
y_offset = img1.shape[0] - 300
roi = img1[y_offset:, x_offset:]

final_img = cv2.bitwise_or(roi, fg)
ax = fig.add_subplot(rows, cols,8)
ax.imshow(final_img)
ax.set_title(f"Blended image")
```

```python
# Threshold method
fig = plt.figure(figsize=(15,15))
rows = 2
cols = 4

def show_pic(num,img,title):
  ax = fig.add_subplot(rows, cols, num)
  ax.imshow(img, cmap='gray')
  ax.set_title(title)


# read an image in grayscale mode
img = cv2.imread('rainbow.jpg', 0)
show_pic(1,img,'read image in grayscale')

# apply fix-level threshold
ret, thresh1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
show_pic(2,thresh1,'apply threshold')

# Inverse Threshold
ret, thresh1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY_INV)
show_pic(3,thresh1,'Inverse threshold')

# Trucate Threshold
ret, thresh1 = cv2.threshold(img, 127, 255, cv2.THRESH_TRUNC)
show_pic(4,thresh1,'Trucate threshold')

# Real world image
img2 = cv2.imread('crossword.jpg',0)
show_pic(5,img2,'Crossword threshold')

# Real world image
ret, th1 = cv2.threshold(img2,180,255, cv2.THRESH_BINARY)
show_pic(6,th1,'apply threshold')

# Automatically find best threshold value
th2 = cv2.adaptiveThreshold(img2, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY,11,8)
show_pic(7,th2, 'Adaptive Threshold')


# Blend achieved threshold to get better result
blended = cv2.addWeighted(src1=th1, alpha=0.6, src2=th2, beta=0.4, gamma=0)
show_pic(8, blended,'Blended Thresholds')
```

```python
# Blurring and smoothing
fig = plt.figure(figsize=(20,10))
rows = 2
cols = 4

def load_img():
  img = cv2.imread('bricks.jpg').astype(np.float32) / 255
  img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
  return img

def display_img(num,img, title):
  ax = fig.add_subplot(rows,cols,num)
  ax.imshow(img)
  ax.set_title(title)

# Original Image
pic = load_img()
display_img(1,pic,'Original image')

# gamma correction
gamma = 8
result = np.power(pic,gamma)
display_img(2,result,'Make darker')
gamma = 1/4
result = np.power(pic,gamma)
display_img(3,result,'Make brighter')

# Blurring 2D convolution, first put Text on image to show blurring
pic = load_img()
font = cv2.FONT_HERSHEY_COMPLEX
cv2.putText(pic, text='bricks', org=(10,600), fontFace=font, fontScale=10, color=(255,0,0), thickness=4)
display_img(4,pic,'Add text')

#Blurring 2D convolution, then create a kernel
kernel = np.ones(shape=(5,5), dtype=np.float32) / 25
dst = cv2.filter2D(pic,-1,kernel)
display_img(5,dst, 'fliter2D Blurring')

# Smoothing based on averages
pic = load_img()
font = cv2.FONT_HERSHEY_COMPLEX
cv2.putText(pic,text='bricks', org=(10,600), fontFace=font, fontScale=10, color=(255,0,0), thickness=4)
blurred = cv2.blur(pic, ksize=(10,10))
display_img(6, blurred, 'Smoothing filter')

# Gaussian blurring
pic = load_img()
font = cv2.FONT_HERSHEY_COMPLEX
cv2.putText(pic,text='bricks', org=(10,600), fontFace=font, fontScale=10, color=(255,0,0), thickness=4)
blurred_img = cv2.GaussianBlur(pic, (5,5), 10)
display_img(7, blurred_img, 'Gaussian blur')

# Median blurring - remove noise
pic = load_img()
font = cv2.FONT_HERSHEY_COMPLEX
cv2.putText(pic,text='bricks', org=(10,600), fontFace=font, fontScale=10, color=(255,0,0), thickness=4)
median_result = cv2.medianBlur(pic, 5)
display_img(8, median_result, 'Median blur')
```
