```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
'''
    Corner Detection
'''
# Prepare the network
fig = plt.figure(figsize=(20,10))
rows = 2
cols = 4
num = 0

def display_img(img,title, cmap=None):
    global num
    num += 1
    ax = fig.add_subplot(rows, cols, num)
    ax.imshow(img, cmap)
    ax.set_title(title)


flat_chess = cv2.imread('flat_chessboard.png')
display_img(flat_chess,'Flat Chessboard')

found, corners = cv2.findChessboardCorners(flat_chess,(7,7))
cv2.drawChessboardCorners(flat_chess,(7,7), corners, found)
display_img(flat_chess,'Grid detection')

dots = cv2.imread('dot_grid.png')
display_img(dots, 'Dot Grid')

found, corners = cv2.findCirclesGrid(dots,(10,10), cv2.CALIB_CB_SYMMETRIC_GRID)
cv2.drawChessboardCorners(dots,(10,10),corners,found)
display_img(dots, 'Grid detection')
```
